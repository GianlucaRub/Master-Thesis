```python
def create_couples(df):
    df = df.drop(columns=['rpctype'],axis = 1).drop_duplicates()
    triplets = []
    for i in range(len(df)):
        head = df.iloc[i]['dm']
        tail = df.iloc[i]['um']
#         rel = df.iloc[i]['rpctype']
        triplets.append([head,tail])
    return triplets
```

```python
def remove_triplet(triplets): # it removes a triplet that has entities that are already seen in other triplets
    
    G = nx.Graph()
    
    services_set= set()
    for triplet in triplets:
        services_set.add(triplet[0])
        services_set.add(triplet[1])
    for elem in services_set:
        G.add_node(elem)
    for triplet in triplets:
        G.add_edge(triplet[0], triplet[1])    
    
    mydict = defaultdict(int)
    for triplet in triplets:
        mydict[triplet[0]]+=1
        if triplet[0] != triplet[1]:
            mydict[triplet[1]]+=1
    
    for triplet in triplets:
        if mydict[triplet[0]]>=2 and mydict[triplet[1]]>=2 and triplet[0] != triplet[1]: #topology based methods do not support self loops
            G.remove_edge(triplet[0],triplet[1]) # the graph needs to remain connected, otherwise topology methods don't work
            if nx.is_connected(G):
                triplets.remove(triplet)
                return triplets, [triplet[0],triplet[1]]
            else:
                G.add_edge(triplet[0],triplet[1])
    return triplets, None
```

```python
def remove_triplets(triplets): # it removes a triplet that has entities that are already seen in other triplets
    
    removed = []
    G = nx.Graph()
    
    services_set= set()
    for triplet in triplets:
        services_set.add(triplet[0])
        services_set.add(triplet[1])
    for elem in services_set:
        G.add_node(elem)
    for triplet in triplets:
        G.add_edge(triplet[0], triplet[1])    
    
    mydict = defaultdict(int)
    for triplet in triplets:
        mydict[triplet[0]]+=1
        if triplet[0] != triplet[1]:
            mydict[triplet[1]]+=1
    
    for triplet in triplets:
        if mydict[triplet[0]]>=2 and mydict[triplet[1]]>=2 and triplet[0] != triplet[1]: #topology based methods do not support self loops
            G.remove_edge(triplet[0],triplet[1]) # the graph needs to remain connected, otherwise topology methods don't work
            if nx.is_connected(G):
                triplets.remove(triplet)
                removed.append([triplet[0],triplet[1]])
            else:
                G.add_edge(triplet[0],triplet[1])
    return triplets, removed
```

```python
def compute_topology_metrics(triplets):
    
    triplets, true_links = remove_triplets(triplets)
    # Create a graph object
    G = nx.Graph()
    
    services_set= set()
    for triplet in triplets:
        services_set.add(triplet[0])
        services_set.add(triplet[1])
    labelsDict = {}
    for elem in services_set:
        G.add_node(elem)
        labelsDict[elem] = elem
    for triplet in triplets:
        G.add_edge(triplet[0], triplet[1], label=triplet[1])
        
#     node_size = 5000
#     font_size = 15
#     bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2')
#     fig = plt.figure(1, figsize=(20, 20), dpi=400)
#     # Use NetworkX's built-in drawing function to visualize the graph
#     pos = nx.spring_layout(G,seed = 545)
#     nx.draw_networkx_edges(G,pos,arrowstyle="-|>",arrowsize=10,width=1,arrows=True,node_size=node_size)
#     nx.draw_networkx_edge_labels(G,pos,edge_labels=nx.get_edge_attributes(G,'label'),font_size = font_size,bbox = bbox,
#                                 verticalalignment = 'center_baseline')

#     nx.draw(G,pos, labels=labelsDict, with_labels = True,node_size=node_size,font_size = font_size,bbox = bbox,node_color='k')
#     plt.show()
#     for triplet in triplets:
#         print(triplet)
        
        
        
    
    non_edges = list(set(nx.non_edges(G)))
#     for elem in set(set(nx.nodes(G)) - set(nx.nodes_with_selfloops(G))):
#         non_edges.append([elem,elem])

#     print(non_edges)
#     print("True link to predict: ",true_link)
#     print()
#     print("Common Neighbours")
    common_neighbours = []

    preds = nx.common_neighbor_centrality(G,ebunch = non_edges, alpha=1)
    for u, v, p in preds:
#         print(f"({u}, {v}) -> {p:.8f}")
        common_neighbours.append(float(p))
    
    
#     print()
#     print("Sorensen Index")
    sorensen_index_list = []
    for elem in non_edges:
        common_neighbors = list(nx.common_neighbors(G, elem[0], elem[1]))
        sorensen_index = (2 * len(common_neighbors)) / (G.degree(elem[0]) + G.degree(elem[1]))
#         print(f"({elem[0]}, {elem[1]}) -> {sorensen_index :.8f}")
        sorensen_index_list.append(sorensen_index)
    
#     print("Salton Index")
    salton_index_list = []
    for elem in non_edges:
        common_neighbors = list(nx.common_neighbors(G, elem[0], elem[1]))
        salton_index = len(common_neighbors) / math.sqrt(G.degree(elem[0]) * G.degree(elem[1]))
#         print(f"({elem[0]}, {elem[1]}) -> {salton_index :.8f}")
        salton_index_list.append(salton_index)
    
#     print()
#     print("Jaccard Index")
    jaccard_index = []
    preds = nx.jaccard_coefficient(G,ebunch = non_edges)
    for u, v, p in preds:
#         print(f"({u}, {v}) -> {p:.8f}")
        jaccard_index.append(float(p))
    
#     print()
#     print("Resouce Allocation")
    resource_allocation = []
    preds = nx.resource_allocation_index(G,ebunch = non_edges)
    for u, v, p in preds:
#         print(f"({u}, {v}) -> {p:.8f}")
        resource_allocation.append(float(p))
    
#     print()
#     print("Adamic Adar")
    adamic_adar = []
    preds = nx.adamic_adar_index(G,ebunch = non_edges)
    for u, v, p in preds:
#         print(f"({u}, {v}) -> {p:.8f}")
        adamic_adar.append(float(p))
        
        
#     print()
#     print("Katz Index")
    katz_index_list = []
    katz_list = katz_index(G,non_edges)
    for elem in zip(non_edges,katz_list):
#         print(f"({elem[0][0]}, {elem[0][1]}) -> {elem[1] :.8f}")
        katz_index_list.append(elem[1])
    
#     print()
#     print("LHN Index")
    lhn_index_list = []
    for elem in zip(non_edges,katz_list):
        lhn_index = elem[1]*2*len(G)/(G.degree(elem[0][0]) * G.degree(elem [0][1]))
#         print(f"({elem[0][0]}, {elem[0][1]}) -> {lhn_index :.8f}")
        lhn_index_list.append(lhn_index)
    return non_edges, common_neighbours, true_links, sorensen_index_list, salton_index_list, jaccard_index, resource_allocation, adamic_adar, katz_index_list, lhn_index_list
```

```python
def katz_index(nx_graph,non_edges):
    to_return = []

    # convert to a Networkit graph
    nk_graph = nk.Graph(len(nx_graph), weighted=True)
    from collections import defaultdict

    # Define a defaultdict with default value as 0
    d = defaultdict(int)
    i = 1
    
    for node in nx_graph.nodes():
        if d[node] == 0:
            i +=1
            d[node] = i
        nk_graph.addNode()
        
    # add edges to Networkit graph
    for edge in nx_graph.edges():
        u, v = edge
        

        nk_graph.addEdge(d[u], d[v])

    katz = nk.linkprediction.KatzIndex(nk_graph)
    
    for elem in non_edges:
        to_return.append(katz.run(d[elem[0]], d[elem[1]]))
    return to_return
```

```python
def calculate_rank(non_edges, metric_list, true_link):
    sorted_data = sorted(zip(non_edges, metric_list),key=lambda x: x[1],reverse = True)

    # Unzip the sorted data into separate lists
    sorted_non_edges, sorted_metric_list = zip(*sorted_data)
    rank = 0
    while rank < len(sorted_non_edges):
        if sorted_non_edges[rank][0] == true_link[0] and sorted_non_edges[rank][1] == true_link[1]:
            break
        elif sorted_non_edges[rank][0] == true_link[1] and sorted_non_edges[rank][1] == true_link[0]:
            break # the graph is undirected
        else:
            rank+=1
    
    return rank +1
```

```python
def calculate_rank_max(non_edges, metric_list, true_link):
    sorted_data = sorted(zip(non_edges, metric_list),key=lambda x: x[1],reverse = True)

    # Unzip the sorted data into separate lists
    sorted_non_edges, sorted_metric_list = zip(*sorted_data)
    rank = 0
    while rank < len(sorted_non_edges):
        if sorted_non_edges[rank][0] == true_link[0] and sorted_non_edges[rank][1] == true_link[1]:
            break
        elif sorted_non_edges[rank][0] == true_link[1] and sorted_non_edges[rank][1] == true_link[0]:
            break # the graph is undirected
        else:
            rank+=1
    
    #using max policy
    i = len(sorted_metric_list)-1
    
    while i >= 0:
        if sorted_metric_list[i] == sorted_metric_list[rank]:
            return i +1
        i+=-1
    return -1
```

```python
def mean_reciprocal_rank(non_edges,metric_list,true_links):
    ranks = []
    for elem in true_links:
        ranks.append(calculate_rank_max(non_edges,metric_list,elem))
        
    return sum(1/x for x in ranks)/len(ranks)
```

```python
def hits_at_k(non_edges,metric_list,true_links,k):
    ranks = []
    for elem in true_links:
        ranks.append(calculate_rank_max(non_edges,metric_list,elem))
    
    hits = 0
    for rank in ranks:
        if rank <k:
            hits+=1
    return hits/len(ranks)
```
