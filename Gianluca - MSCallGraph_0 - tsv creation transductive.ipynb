```python
import pandas as pd
import numpy as np

seed = 1234
```

```python
df = pd.read_csv('MSCallGraph_0.csv')
df = df.drop(['Unnamed: 0','traceid','timestamp','rpcid','interface','rt',],axis=1)
df = df.replace('(?)', np.NaN)
df = df.replace('', np.NaN)
df = df.replace('NAN', np.NaN)
df = df.dropna()
df = df.drop_duplicates()
```

```python
df
```

```python
test_df = df
```

```python
entities = set(df['dm']) | set(df['um'])
len(entities)
```

```python
baseline = []
for entity in entities:
    baseline.append(df[df['dm'] == entity].head(1))
    baseline.append(df[df['um'] == entity].head(1))
```

```python
baseline_df = pd.concat(baseline)
baseline_df = baseline_df.drop_duplicates()
baseline_df
```

```python
n_train_items = int(len(df)*0.64)
n_items_to_add = n_train_items - len(baseline_df)
print("the number of items to add to the train set is: ",n_items_to_add)
```

```python
not_baseline_df = df.merge(baseline_df, how='outer', indicator=True).loc[lambda x : x['_merge']=='left_only'].drop('_merge',axis= 1)
not_baseline_df
```

```python
train_df = pd.concat([baseline_df, not_baseline_df.sample(n_items_to_add,random_state = seed)])
train_df
```

Transductive checks

```python
assert(len(set(train_df['dm']) or set(train_df['um'])) == len(set(test_df['dm']) or set(test_df['um'])))
```

```python
assert len(set(train_df['dm'])) == len(set(test_df['dm']))
```

```python
assert len(set(train_df['um'])) == len(set(test_df['um']))
```

```python
assert(len(set(train_df['rpctype']))== len(set(test_df['rpctype'])))
```

```python
remaining_df = df.merge(train_df, how='outer', indicator=True).loc[lambda x : x['_merge']=='left_only'].drop('_merge',axis= 1)
```

```python
validation_df = pd.concat([train_df,remaining_df.sample(frac = 0.44, random_state = seed)])
validation_df
```

Check to see if there is test leakage

```python
assert(len(train_df)==len(pd.merge(train_df, validation_df, how='inner')))
```

```python
assert(len(validation_df)==len(pd.merge(test_df, validation_df, how='inner')))
```

```python
def create_triplets(df):
    triplets = []
    for i in range(len(df)):
        head = df.iloc[i]['dm']
        tail = df.iloc[i]['um']
        rel = df.iloc[i]['rpctype']
        triplets.append([head,rel,tail])
    return triplets
```

```python
import asposecells
import jpype
jpype.startJVM()

from asposecells.api import Workbook, FileFormatType

def create_tsv(triplets, file_name):


    # Create Workbook object.
    workbook = Workbook(FileFormatType.TSV)

    # Access the first worksheet of the workbook.
    worksheet = workbook.getWorksheets().get(0)

    # Get the desired cell(s) of the worksheet and input the value into the cell(s).


    i = 1
    for elem in triplets: 
        worksheet.getCells().get("A"+str(i)).putValue(elem[0])
        worksheet.getCells().get("B"+str(i)).putValue(elem[1])
        worksheet.getCells().get("C"+str(i)).putValue(elem[2])
        i+=1


    # Save the workbook as TSV file.
    workbook.save(file_name)


    file = open(file_name,'r')  
    lines = file.readlines()  
    file.close()
    file = open(file_name,'w')  
    lines = lines[:-1]
    file.writelines(lines)
    file.close()
```

```python
datasets = [train_df, validation_df, test_df]
names = ["train", "validation","test"]
```

```python
for elem in zip(datasets,names):
    create_tsv(create_triplets(elem[0]),f"MSCallGraph_0_{elem[1]}_transductive.tsv")
jpype.shutdownJVM()
```
